```python
df[:3]
```

```python
def custom_format(x):
    if x >= 1e9:
        return f"{x / 1e9:.1f} B"  # Billions
    elif x >= 1e6:
        return f"{x / 1e6:.1f} M"  # Millions
    elif x >= 1e3:
        return f"{x / 1e3:.1f} K"  # Thousands (optional)
    else:
        return f"{x:,.0f}"  # Default (no suffix)
```

```python
# df.groupby("Continent")["Total_income"].mean().sort_values(ascending=False)


df.groupby("Continent")["Total_income"].sum().sort_values(ascending=False).apply(custom_format)

# df.groupby("Continent", group_keys=True).CountryID.count()
```

```python
df.groupby("Continent")["Total_income"].agg(
    ["count", "mean", "sum", "std"]
).style.format(
    {
        "count": "{:.0f}",  # Keep count as integer
        "mean": custom_format,  # Format mean as B/M
        "sum": custom_format,  # Format sum as B/M
        "std": custom_format,  # Keep std as decimal
    }
)
```
