```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://drive.google.com/uc?id=1oE-3rt17bFW7fOzDIjwFSEMPTIV3NvcO")
```

```python
df[df["grade_math_t2"] > 10]
```

```python

df.dropna(inplace=True)


math_t2_mean = df[df["grade_math_t2"] <= 10]["grade_math_t2"].mean()
lang_t2_mean = df[df["grade_language_t2"] <= 10]["grade_language_t2"].mean()
science_t2_mean = df[df["grade_science_t2"] <= 10]["grade_science_t2"].mean()

df["grade_math_t2"] = df["grade_math_t2"].where(
    df["grade_math_t2"] <= 10, other=math_t2_mean
)
df["grade_language_t2"] = df["grade_language_t2"].where(
    df["grade_language_t2"] <= 10, other=lang_t2_mean
)
df["grade_science_t2"] = df["grade_science_t2"].where(
    df["grade_science_t2"] <= 10, other=science_t2_mean
)

df.describe()
```

```python
df.groupby("grade")["grade_math_t1"].mean()
df.groupby("grade", dropna=False)["grade_math_t2"].sum()

# When using .apply(), use group_keys to include or exclude the group keys.
df.groupby("grade", group_keys=False)["grade_language_t1"].apply(lambda x: x.mean()+2)

df.groupby("grade", dropna=False)["grade_math_t2"].count()
```

```python
df.groupby("grade")["grade_language_t2"].agg(["count", "mean", "sum", "std"])
```
