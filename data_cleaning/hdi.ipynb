```python
import pandas as pd
```

```python
df = pd.read_csv('../Data/human_development_index.csv', encoding='latin-1')

df[:3]
```

```python
df_hdi = df.melt(
    id_vars=['iso3', 'country', 'hdicode', 'region', 'hdi_rank_2022'],
    value_vars=[col for col in df.columns if col.startswith('hdi_')],
    var_name='year',
    value_name='hdi',
)
df_hdi[:2]
```

```python
df_hdi['year'] = df_hdi['year'].str.replace('hdi_', '')
df_hdi[:2]
```

```python
df_pop = df.melt(
    id_vars=['iso3', 'country', 'hdicode', 'region', 'hdi_rank_2022'],
    value_vars=[col for col in df.columns if col.startswith('pop_total_')],
    var_name='year',
    value_name='pop_total',
)

df_pop[:3]
```

```python
df_pop['year'] = df_pop['year'].str.replace('pop_total_', '')
df_pop[:3]
```

```python
df_merge = df_hdi.merge(
    df_pop,
    how='left',
    on=['iso3', 'country', 'hdicode', 'region', 'hdi_rank_2022', 'year'],
)
df_merge[:3]
```
