```python
import csv

def simplify_csv(input_file, output_file):
    with open(input_file, mode='r', encoding='utf-8') as infile, \
         open(output_file, mode='w', encoding='utf-8', newline='') as outfile:

        reader = csv.DictReader(infile)
        writer = csv.writer(outfile)

        # Write the header for the new CSV file
        writer.writerow(['Location', 'Track #', 'Key'])

        # Write the rows with only the necessary columns
        for row in reader:
            writer.writerow([row['Location'], row['Track #'], row['Key']])

# Example usage
input_file = '../Data/01_key_analyzed.csv'
output_file = '../Data/01_key_analyzed_simple.csv'
simplify_csv(input_file, output_file)
```
