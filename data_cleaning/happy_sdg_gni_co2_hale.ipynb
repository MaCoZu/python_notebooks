```python
happy.columns = ['country', 'code', 'year', 'happiness score']
happy[:3]
```

```python
co2.columns = ['country', 'year', 'annual CO2 emission pc. (in tonnes)']
co2_2011 = co2[co2.year >= 2011]
co2_2011[:3]
```

```python
happy_co2 = happy.merge(co2_2011, on=['country', 'year'], how='outer')
happy_co2[:5]
```

```python
hdi_2011 = hdi[hdi.year >= 2011]
hdi_2011[:3]
```

```python
happy_co2_hdi = happy_co2.merge(hdi_2011, on=['country', 'year'], how='outer')
happy_co2_hdi[:4]
```

```python
sdg_2011 = sdg[sdg.year >=2011]
sdg_2011.year.unique()
```

```python
happy_co2_hdi_sdg = happy_co2_hdi.merge(sdg_2011, on=['country', 'year'], how='outer')
happy_co2_hdi_sdg[300:305]
```

```python
#  consolidate iso/country columns
happy_co2_hdi_sdg['country_code'] = (
    happy_co2_hdi_sdg['code']
    .combine_first(happy_co2_hdi_sdg['iso'])
    .combine_first(happy_co2_hdi_sdg['iso3'])
)

happy_co2_hdi_sdg[200:205]
```

```python
happy_co2_hdi_sdg_redux = happy_co2_hdi_sdg.drop(
    columns=['code', 'iso', 'iso3', 'hdi_rank_2022']
)
happy_co2_hdi_sdg_redux[400:403]
```

```python
# creating bool series True for NaN values
hchsr = happy_co2_hdi_sdg_redux.copy()
null_series = pd.isnull(hchsr['happiness score'])
hchsr[null_series][:5]
```

```python
hchsr.isnull().sum()
```

```python
hchsr_nonnull = hchsr.dropna(subset=['happiness score'])
hchsr_nonnull.isnull().sum()
```

```python
hchsr_nonnull_redux = hchsr_nonnull[
    [
        'country',
        'country_code',
        'year',
        'happiness score',
        'hdi',
        'sdg_index',
        'annual CO2 emission pc. (in tonnes)',
        'pop_total',
    ]
]
hchsr_nonnull_redux
```

```python
hchsr_nonnull_redux.to_csv('../Data/happy_sdg_hdi_co2_clean.csv', index=False)
```
