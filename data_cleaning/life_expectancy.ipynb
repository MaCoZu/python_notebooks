```python
import pandas as pd
```

```python
df = pd.read_csv('../Data/Life Expectancy.csv')
df[:3]
```

```python
df.info()
```

```python
df_select = df[['ParentLocationCode', 'ParentLocation', 'Location', 'Period', 'Value']]
df_select[:4]
```

```python
df_order = df_select[
    ['Location', 'ParentLocation', 'ParentLocationCode', 'Period', 'Value']
]
df_order.columns = ['Country', 'Region', 'Country Code', 'Year', 'Life_Expectancy']
df_order[:4]
```

```python
df_order[['Life_Expectancy Number', 'Life_Expectancy Range']] = (
    df_order['Life_Expectancy']
    .str.extract(r'^(\d+\.\d+)\s*\[([\d\.]+-[\d\.]+)\]')
    .copy()
)

df_order[:3]
```

```python
df_order.drop(columns=['Life_Expectancy'], inplace=True)
```

```python
df_order[:3]
df_order.to_csv('../Data/life_expectancy_clean.csv', index=False)
```
