# concat
- Like its sibling function on ndarrays, numpy.concatenate, pandas.concat takes a list or dict <br>
of homogeneously-typed objects and concatenates them with some configurable <br>
handling of what to do with the other axes
- concat() (and therefore append()) makes a full copy of the data, and that constantly reusing <br>
this function can create a significant performance hit. 
-  If you need to use the operation over several datasets, **use a list comprehension**.
- When gluing together multiple DataFrames, you have a choice of how to handle the other axes <br>
(other than the one being concatenated). This can be done in the following two ways:<br>
  - Take the union of them all, **join='outer'**. This is the default option as it results in zero information loss.
  - Take the intersection, **join='inner'**.


```python
df1 = pd.DataFrame(
     {
         "A": ["A0", "A1", "A2", "A3"],
         "B": ["B0", "B1", "B2", "B3"],
         "C": ["C0", "C1", "C2", "C3"],
         "D": ["D0", "D1", "D2", "D3"],
     },
     index=[0, 1, 2, 3],
 )


df2 = pd.DataFrame(
     {
         "A": ["A4", "A5", "A6", "A7"],
         "B": ["B4", "B5", "B6", "B7"],
         "C": ["C4", "C5", "C6", "C7"],
         "D": ["D4", "D5", "D6", "D7"],
     },
     index=[4, 5, 6, 7],
 )


frames = [df1, df2]
```

## hierarchical index

# merge
-  merge(), is the entry point for all standard database join operations
- pandas has in-memory join operations very similar to relational databases like SQL
- In Excel, merging of tables can be done through a VLOOKUP. 
- The data does not have to be sorted ahead of time, and different join types are <br>
accomplished via the **how** keyword.<br>
- The related join() method, uses merge internally for the index-on-index (by default) and <br>
column(s)-on-index join. If you are joining on index only, you may wish to use df.join to save yourself some typing.


Join operations between two SQL-table like structures 
- **one-to-one** joins: for example when joining two df's on their indexes (which must contain unique values).
- **many-to-one** joins: for example when joining an index (unique) to one or more columns in a different df
- **many-to-many** joins: joining columns on columns.



```python
df1 = pd.DataFrame({"key": ["A", "B", "C", "D"], "value": np.random.randn(4)})
df2 = pd.DataFrame({"key": ["B", "D", "D", "E"], "value": np.random.randn(4)})
```

```python
df1
```

## cross / cartesian join

```python
# Create the cartesian product of rows of both frames
cross = df1.merge(df2, how='cross')
cross
```

## inner join

```python
# only where both match, inner is the default
inner_join = df1.merge(df2, on=["key"])
inner_join
```

## left join

```python
left_join = df1.merge(df2, on=["key"], how="left") # all rows from the left table
left_join
```

## right join

```python
right_join = df1.merge(df2, on=["key"], how="right") # all from the right
right_join
```

## outer 

```python
outer_join = df1.merge(df2, on=["key"], how="outer") # all rows from both tables
outer_join
```

## merge on multiple keys
- Only the keys appearing in left and right are present (the intersection), since how='inner' by default.

```python
left = pd.DataFrame({
    "key1": ["K0", "K0", "K1", "K2"],
    "key2": ["K0", "K1", "K0", "K1"],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"],
})

right = pd.DataFrame({
    "key1": ["K0", "K1", "K1", "K2"],
    "key2": ["K0", "K0", "K0", "K0"],
    "C": ["C0", "C1", "C2", "C3"],
    "D": ["D0", "D1", "D2", "D3"],
})

left
```

```python
pd.merge(left, right, on=["key1", "key2"])
```

```python
# The how argument to merge specifies how to determine which keys are to be included in the 
# resulting table. If a key combination does not appear in either the left or right tables, 
# the values in the joined table will be NA.
pd.merge(left, right, how="right", on=["key1", "key2"])
pd.merge(left, right, how="left", on=["key1", "key2"])
pd.merge(left, right, how="outer", on=["key1", "key2"])
```

# pivot tables
- **index**: Column to use to make new frames index. If None, uses existing index.
- **columns**: Column to use to make new frames columns.
- **values**: Column(s) to use for populating new frames values. <br>
If not specified, all remaining columns will be used and the result will have hierarchically indexed columns.

```python
df = pd.DataFrame({
    "ix": ['one'] * 6 + ['two'] * 6,
    "cat": ['A', 'B', 'C'] * 4,
    "val": list(range(1, 7))*2,
    "val2": list(range(10, 22)),
    "cat2":['x', 'y', 'z', 'q', 'w', 't']*2
})

df
```

```python
# Create a (pivot) table that aggregates values in one column ('val') which 
# have the same index ('ix') but differentiates them by a second column ('cat)
df.pivot_table(index="ix", columns='cat', values=["val", 'val2'], aggfunc=sum)
```

```python
# function for grouping countries by ineqality index
def inequality_groups(x):
    if x <= 20:
        return '0-20'
    elif 20 < x <=40:
        return '20-40'
    elif 40 < x <=60:
        return '40-60'
    elif 60 < x <=80:
        return '60-80'
    elif 80 < x <=100:
        return '80-100'
    else: np.nan

# apply function row by row
selection['inequality_group'] = selection['Inequality_index'].apply(lambda row: inequality_groups(row))

# rearrange columns
selection = selection[[
    'Continent',
    'Total_CO2_emissions',
    'Income_per_person',
    'Inequality_index',
    'inequality_group',
    'Population_total',
    'Life expectancy at birth (years) both sexes',
    'Personal_computers_per_100_people',
    'Patent_applications',
    'Military_expenditure',
]]

# rename columns
selection.columns = [
    'Continent_number',
    'Total_CO2_emissions',
    'Income_pc',
    'Inequality_index',
    'inequality_group',
    'pop_total',
    'life_expectancy',
    'computer_per_100_people',
    'Patent_applications',
    'Military_expenditure'
]


# dict map numbers to actual names
Continents = {
    1: "Eastern Mediterranean",
    2: "Europe",
    3: "Africa",
    4: "Americas",
    5: "South America",
    6: "South-East Asia",
    7: "South & Eastern Asia"
}

# map names to numbers and insert coumn at position 2
selection.insert(1, 'Continent', selection.Continent_number.map(Continents))
selection.head(3)
```

```python
piv = pd.pivot_table(selection,
               values=["Income_pc"],
               index=["Continent"],
               columns='inequality_group',
               aggfunc=np.average)

piv.round(0).astype(pd.Int64Dtype()) # rounding and converting to int
```

## merge
- The join is done on columns or indexes. 
- If joining columns on columns, the DataFrame indexes will be ignored. 
- Otherwise if joining indexes on indexes or indexes on a column or columns, the index will be passed on. 
- When performing a cross merge, no column specifications to merge on are allowed.

```python
merged_df = pd.merge(df,df1) # Merge on common columns
df2 = pd.merge(df, df1, on=['Courses','Fee']) # Use pandas.merge() on multiple columns
df2 = pd.merge(df, df1,  how='left', left_on=['Courses','Fee'], right_on = ['Courses','Fee']) # Use pandas.merge() on multiple columns
merged_df = pd.merge(df, df1, left_on="Courses", right_on="Courses") # Merge Pandas DataFrames using left_on and right_on
merged_df = pd.merge(df, df1, on="Courses") # Set value of on parameter to specify the key value for merge in pandas
```
