# Boolean Indexing, Sorting

```python
import pandas as pd
```

```python
df = pd.read_csv("https://drive.google.com/uc?id=1oE-3rt17bFW7fOzDIjwFSEMPTIV3NvcO")
```

## Boolean Indexing

```python
df[df["school_id"] == 57] # filter rows by column value
df[df["school_id"] != 57] # filter rows by column value


df[df["school_id"].isin([57, 141])] # multiple values in a column
df[df["date_of_birth"]>"1995-01-01"] # greater than


df[df["grade_language_t1"].between(8.6, 8.7)] # between function


df[(df["grade_math_t1"] > 9) | (df["school_id"] == 57)] # or
df[(df["grade_math_t1"] > 9) & (df["school_id"] == 57)] # and


df[~(df["grade_math_t1"] < 8)] # negate - NOT less than 8
df[~df["school_id"].isin([57, 141])] # negate - NOT in list of values


# boolean indexing and choosing columns
df[df.grade_language_t1 > 9][["school_id", "grade"]]
```

### query()

```python
df[df.grade_math_t1 > df.grade_math_t2]


df.query("grade_math_t1 > grade_math_t2")


# query can be cleaner for complex queries - keyword class needs backticks in query
df[(df.sex == 1) & (df["class"] == "C") & ((df.grade_math_t1 > 9.9) | (df.grade_math_t2 > 9.9))]
df.query("sex == 1 & `class` == 'C' & (grade_math_t1 > 9.9 | grade_math_t2 > 9.9)")
```

```python
df.query("grade_math_t1 > grade_math_t2")[["grade_math_t1", "grade_math_t2"]].round(2)
```

### query in sql for comparison

```python
from pandasql import sqldf

# Define a helper function for SQL queries
def query(q):
    return sqldf(q, globals())
```

```python
result = query("""
    SELECT 
        ROUND(grade_math_t1) as grade_math_t1, 
        ROUND(grade_math_t2) as grade_math_t2,
    FROM df
    WHERE grade_math_t1 > grade_math_t2
""")

result
```

## Sorting

```python
df.sort_values(by="school_id", ascending=True) # sort by column
df.sort_values(by=["school_id", "grade_language_t1"], ascending=[True, False]) # sort by multiple columns

df.sort_index() # sort by index
df.sort_index(ascending=False) # sort by index in descending order
```

### nlargest, nsmallest, rank

```python
df.nlargest(5, "grade_language_t1") # top 5 rows with largest values in column
df.nsmallest(5, "grade_language_t1") # top 5 rows with smallest values in column


# add a new column with rank based on another column
df["rank"] = df["grade_language_t1"].rank(ascending=False, method='dense') 

df.sort_values(by="rank")
```

### value_counts() 
for categorical columns
