## How to Create a DataFrame?
A DataFrame which is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure. There are several ways to create a DataFrame in pandas. Here are the most common ones:

* From a dictionary of lists
* From a list of lists
* From CSV or other files

#### Example
Let us explore the above-mentioned ways through examples.

* From a dictionary of lists or arrays

````python
# Creating a DataFrame from a dictionary of lists
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Paris', 'Berlin']
}
df = pd.DataFrame(data)
print(df)
````

* From a list of lists

````python
# Creating a DataFrame from a list of lists
data = [
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'Paris'],
    ['Charlie', 35, 'Berlin']
]
df = pd.DataFrame(data, columns=['Name', 'Age', 'City'])
print(df)
````

* From CSV or other files:
  * .csv or .xlsx files
  * Urls where the dataset is residing

````python
# Read data for car prices
url = (
    "https://raw.githubusercontent.com/juliandnl/redi_ss20/master/cars.csv"
 )
df = pd.read_csv(url)
df.head()
````

## Pandas vs SQL

Pandas and SQL both are commonly used for handling and querying data. While SQL is a database language to work with relational databases, Pandas is a Python library that allows you to work with data in DataFrames, which are like in-memory tables.

Operation | SQL | Pandas
:-: | :-: | :-:
Select columns | `SELECT column_name FROM table_name` | `df['column_name']`
Filter rows	 | `SELECT * FROM table WHERE condition` | `df[df['column_name'] == condition]`
Aggregate functions | `GROUP BY` |  `df.groupby(['column']).agg()`
Join tables | `JOIN` |  `pd.merge(df1, df2, on='column')`
