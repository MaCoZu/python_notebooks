```python
import pandas as pd
```

```python
df = pd.read_csv("https://drive.google.com/uc?id=1oE-3rt17bFW7fOzDIjwFSEMPTIV3NvcO")
```

## unique & nunique

```python
# What are the unique grades in school 57 and 946?
df[df["school_id"]==57]["grade"].unique()
df[df["school_id"]==946]["grade"].unique()

# How many unique grades are there in school 57 and 946?
df[df["school_id"]==946]["grade"].nunique()
df[df["school_id"]==946]["grade"].nunique()
```

```python
df.groupby("grade")["grade_math_t1"].sum()
df.groupby("grade")["grade_math_t1"].count()
df.groupby("grade")["grade_math_t1"].min()
df.groupby("grade")["grade_math_t1"].max()
df.groupby("grade")["grade_math_t1"].std()
df.groupby("grade")["grade_math_t1"].var()
df.groupby("grade")["grade_math_t1"].median()
df.groupby("grade")["grade_math_t1"].mean()
```

```python
# agg() function - allows us to apply multiple aggregation functions at once
df.groupby("grade")["grade_math_t1"].agg(["count", "sum", "mean", "min", "max"])

# Renaming the columns
df.groupby("grade")["grade_math_t1"].agg(
    Average = "mean",
    Total   = "sum",
    Lowest  = "min",
    Highest = "max"
)
```

```python
# Custom aggregation function possible but not vectorized 
df.groupby("grade")["grade_math_t1"].agg((lambda x: sum(x) / len(x)))
```

## apply()

```python
# apply() function - designed for flexible, custom operations (e.g., row-wise or group-wise logic).
df.groupby("grade")["grade_math_t1"].apply((lambda x: sum(x) / len(x)))
```

## transform()

```python
# transform preserves original df shape, but allows us to apply groupby aggregation functions
df["grade_math_t1_mean"] = df.groupby("grade")["grade_math_t1"].transform("mean")

df[["grade_math_t1", "grade", "grade_math_t1_mean"]].sample(10).sort_values("grade")
```

## filter()

```python
# same as boolean indexing but more flexible
df.groupby("grade").filter(lambda x: x["grade_math_t1"].mean() > 6.7)

# column to check filter statement 
df["sex_math_t1_mean"] = df.groupby("sex")["grade_math_t1"].transform("mean")        

# filter groups based on the mean of grade_math_t1 for each
df.groupby("sex").filter(lambda x: x["grade_math_t1"].mean() > 6.7).sample(10)

# regex filter
df.filter(regex="^grade", axis=1)

# fuzzy filter
df.filter(like="math")
```

## size(), count(), nunique()

```python

df.groupby("class").size()    # counts ALL rows per group, including NaN values
df.groupby("class").count()   # counts only NON-NULL values per column per group
```

```python
# groupby multiple columns - use parenethese for chained operations
counted_df = (
    df.groupby(["school_id", "class"])
    ["grade"]
    .count()
    .reset_index() 
    .query("school_id == 57")
)

counted_df
```

```python
# How many unique products does each country sell?
df.groupby("school_id")["class"].nunique()
```

## first() and last() functions

```python
df.groupby("sex").first()    # first row of each sex group
df.groupby("school_id").last()     # last row of each school_id group
```

## isin() & ~exclude

```python
df[df["school_id"].isin([57, 262, 946])]    # keep these schools
df[~df["school_id"].isin([57, 262, 946])]    # exclude these schools
```

```python
df.query("price > 900")
df.query("gender == 'Female'")
df.query("price > 900 and gender == 'Female'")
df.query("country in ['USA', 'UK']")
```

```python
df["price"].where(df["price"] > 900)           # keep prices > 900, rest → NaN
df["price"].where(df["price"] > 900, other=0)  # keep prices > 900, rest → 0


df.set_index("country")                        # use country as the index

# inplace=False (default) — original is unchanged, result is a new DataFrame
df2 = df.set_index("country")
# inplace=True — original DataFrame is modified directly
df.set_index("country", inplace=True)



# After grouping by country and gender
grouped = df.groupby(["country", "gender"])["price"].mean().reset_index()
grouped = grouped.set_index(["country", "gender"])
# Select all rows where gender == "Female" across all countries
grouped.xs("Female", level="gender")


# Average price per country, then select only the top 2
df.groupby("country")["price"].mean().nlargest(2)


# Flag rows where price exceeds the group average for their gender
df["gender_avg"] = df.groupby("gender")["price"].transform("mean")
df[df["price"] > df["gender_avg"]]


# Total and average price per country, then keep only countries with avg > 800
result = df.groupby("country")["price"].agg(Total="sum", Average="mean")
result.query("Average > 800")
```
