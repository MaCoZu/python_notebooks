# Selection, Indexing, Filtering

### sorting

```python
df.sort_values(by="school_id", ascending=True) # sort by column
df.sort_values(by=["school_id", "grade_language_t1"], ascending=[True, False]) # sort by multiple columns

df.sort_index() # sort by index
df.sort_index(ascending=False) # sort by index in descending order
```

### nlargest, nsmallest, rank

```python
df.nlargest(5, "grade_language_t1") # top 5 rows with largest values in column
df.nsmallest(5, "grade_language_t1") # top 5 rows with smallest values in column

df["rank"] = df["grade_language_t1"].rank(ascending=False, method='dense') # add a new column with rank based on another column
df.sort_values(by="rank")
```

```python
df.sample(3) # random sample of 10 rows
```

```python
df.value_counts("school_id") # count unique values in a column
```

## Boolean Indexing

```python
df[df["school_id"] == 57] # filter rows by column value
df[df["school_id"].isin([57, 141])] # multiple values in a column
df[df["date_of_birth"]>"1995-01-01"] # greater than
df[(df["grade_math_t1"] > 9) | (df["school_id"] == 57)] # or
df[(df["grade_math_t1"] > 9) & (df["school_id"] == 57)] # and

# boolean indexing and choosing columns
df[df.grade_language_t1 > 9][["school_id", "grade"]]
```

### between()

```python
# between function
df[df["grade_language_t1"].between(8.6, 8.7)]
```

### isin()

```python
# if values is a list check whether every value in the DataFrame is present in the list
df[df.isin([57, 0])] 

# check specific column for multiple values
df[df.grade.isin([6, 7])]

# check specific column for multiple values and negate the condition
df[~df.grade.isin([6, 7])]
```

### query()

```python
df[df.grade_math_t1 > df.grade_math_t2]
df.query("grade_math_t1 > grade_math_t2")[["grade_math_t1", "grade_math_t2"]].round(2)
```

### query in sql for comparison

```python
from pandasql import sqldf

# Define a helper function for SQL queries
def query(q):
    return sqldf(q, globals())
```

```python
result = query("""
    SELECT 
        ROUND(grade_math_t1) as grade_math_t1, 
        ROUND(grade_math_t2) as grade_math_t2,
    FROM df
    WHERE grade_math_t1 > grade_math_t2
""")

result
```

### xs() 
Return cross-section from a particular level of a MultiIndex.

```python
d = {
     "num_legs": [4, 4, 2, 2],
     "num_wings": [0, 0, 2, 2],
     "class": ["mammal", "mammal", "mammal", "bird"],
     "animal": ["cat", "dog", "bat", "penguin"],
     "locomotion": ["walks", "walks", "flies", "walks"],
 }

df = pd.DataFrame(data=d)
df = df.set_index(["class", "animal", "locomotion"])
df
```

```python
# three index levels
df.sort_index(inplace=True)  # sort by first index level
df
```

```python
df.xs("mammal") # get all mammals
df.xs(("mammal", "dog")) # get all info about dogs
df.xs(("mammal", "bat")) # get specific info about dogs
```

```python

df.xs("cat", level=1) # get all animals that are cats
df.xs("walks", level=2) # get all animals that walk
```

```python

# axis=1 means columns
df.xs("num_wings", axis=1) # returns this column


df.xs("mammal", level=0) # level of the index
df.xs("dog", level="animal")

# multiple levels one by index and one by name
df.xs(("bird", "walks"), level=[0, "locomotion"])
```
