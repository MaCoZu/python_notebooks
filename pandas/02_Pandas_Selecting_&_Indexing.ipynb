###  filter()

```python
# alternative with extra functionality
df.filter(["sex", "grade_language_t1"])
df.filter(items=["sex", "grade"]) # select columns

# regex filtering
df.filter(regex="^grade", axis=1) # select columns with "grade" at the start of their name
df.filter(regex="t2$", axis=1) # select columns with "t2" at the end

# fuzzy filtering
df.filter(regex="^grade") # all columns that start with 'grade'
df.filter(like="math", axis=1) # select columns with "math" in their name
```

## Index Rows
accessing specific positions (rows/cells)

```python
# create a new dataframe with a string index
df_name = df[["student_id", "grade_language_t1", "school_id"]].set_index("student_id") # set index to a column
df_name.head()

# loc[] selects by name
df_name.loc["j0ihe@cunb.edu"][['school_id', 'grade_language_t1']] # select row by name
```

```python
# iloc[] selects by index
df.iloc[1509][:2]  # happens to the same as name
```

```python
# indexing by row/col name 
df_name.at["j0ihe@cunb.edu", "school_id"]  = 9999  # set value of single cell
df_name.at["j0ihe@cunb.edu", "school_id"]  # single cell

df_name["j0ihe@cunb.edu":].head()
```

```python
# indexing by row/col position
df.iat[1502, 3] = "eohgp-ws@cunb.edu" # update value of single cell
df.iat[1502, 3] # row 1509, col 3 (email)

df[1500:1505]
```

```python
df.loc[200:204]  # select rows by name sliced -> from : to
df[200:204]  # select rows by index sliced -> from : to

# names in the index
df_id = df.set_index("student_id", drop=True)
df_id.loc[['j0ihe@cunb.edu', 'mzmqb@cunb.edu']]
# df_id.iloc[['j0ihe@cunb.edu', 'mzmqb@cunb.edu']] # does not work because iloc only accepts integer positions

df_id[1:4]
df_id.iloc[1:4]
```

### Slicing
Slicing means getting a range of rows using position

```python
# slicing rows, columns


# slicing rows and columns with loc
df.loc[:2, ["sex", "school_id"]]  # slice of rows and selection of columns
df.loc[:, "sex": "grade_language_t1"]  # get all row from col 'b' & 'c'
df.loc[1:3, "sex": "grade_language_t1"]  # get rows from : to, of columns from:to
df.loc[1:2, :]  # get rows - from : to & all columns


# slicing rows and columns with iloc
df.iloc[0:3]  # first 3 rows by index
df.iloc[:, -1]  # select last column by index
df.iloc[0:3, :]  #  Slicing Rows and Columns by position
df.iloc[:, 0:3]  # slice columns by index position.
df.iloc[0:2, 1:3]  # slice row and columns by index position.
df.iloc[:2, :2]  # row index 0 to index 1 (exclusive 2), column zero to one

df.iloc[0:20:4, 0:4]  # row index 0 to 20 (step 4), column index 0 to 4
df.iloc[10::-1] # reverse the order of rows from index 10 to the beginning
```
